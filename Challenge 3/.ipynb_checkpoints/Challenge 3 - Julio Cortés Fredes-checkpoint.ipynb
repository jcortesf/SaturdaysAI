### 5.- ¿Cuáles variables tienen más del 50% de valores NaN?

```python
for i in df.columns:
    if df[i].isnull().sum() > (len(df)/2):
        print(i+': '+ str(df[i].isnull().sum()))
```

### 6.- ¿Cuántos valores faltantes tiene cada una de las variables categóricas?

```python
dfCategorycal = df.select_dtypes(include=['object'])
for i in dfCategorycal.columns:
    if df[i].isnull().sum() > 0:
        print(i+': '+ str(df[i].isnull().sum()))
```

### 7.- ¿Cuántos valores faltantes tiene cada una de las variables númericas?


```python
dfNumeric = df.select_dtypes(include=['int','float'])
for i in dfNumeric.columns:
    if df[i].isnull().sum() > 0:
        print(i+': '+ str(df[i].isnull().sum()))
```

### 8.- Sustituye los valores faltantes de la columna YrSold con la media de esa columna


```python
# Imputación de acuerdo a la media
df['YrSold'] = df['YrSold'].fillna(value=df['YrSold'].mean())
df['YrSold'].head()
```

### 9.- Sustituye los valores faltantes de la columna LotFrontage con la mediana de esa columna

```python
# Imputación de acuerdo a la media
df['LotFrontage'] = df['LotFrontage'].fillna(value=df['LotFrontage'].median())
df['LotFrontage'].head()
```
